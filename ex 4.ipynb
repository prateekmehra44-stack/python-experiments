```python
#Question 1:

num = int(input("Enter number: "))
fact = 1

for i in range(1, num + 1):
    fact = fact * i

print("Factorial:", fact)
```

```python
#Question 2:

num = int(input("Enter number: "))
temp = num
sum = 0
n = len(str(num))

while temp > 0:
    digit = temp % 10
    sum = sum + digit ** n
    temp = temp // 10

if sum == num:
    print("Armstrong Number")
else:
    print("Not Armstrong Number")
```

```python
#Question 3:

n = int(input("Enter number of terms: "))

a = 0
b = 1

for i in range(n):
    print(a, end=" ")
    c = a + b
    a = b
    b = c
```

```python
#Question 4:

num = int(input("Enter number: "))
count = 0

for i in range(1, num + 1):
    if num % i == 0:
        count += 1

if count == 2:
    print("Prime Number")
else:
    print("Not Prime Number")
```

```python
#Question 5:

num = int(input("Enter number: "))
temp = num
rev = 0

while temp > 0:
    digit = temp % 10
    rev = rev * 10 + digit
    temp = temp // 10

if rev == num:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
#Question 6:

num = int(input("Enter number: "))
sum = 0

while num > 0:
    digit = num % 10
    sum = sum + digit
    num = num // 10

print("Sum of digits:", sum)
```

```python
#Question 7:

count = 0

for i in range(1, 101):
    if i % 5 == 0 or i % 7 == 0:
        print(i, end=" ")
        count += 1

print("\nTotal count:", count)
```

```python
#Question 9:

for num in range(1, 101):
    count = 0
    for i in range(1, num + 1):
        if num % i == 0:
            count += 1
    if count == 2:
        print(num, end=" ")
```
